Finally, we will calculate the linear regression between the price and the weight. This will be done first by plotting it with the `seaborn` library and then calculating the error with the `scipy` library.

```python
from scipy import stats

# Assuming you have already loaded the diamonds data into a DataFrame named 'diamonds'
# For example: diamonds = pd.read_csv('diamonds.csv')

# Plotting linear regression between price and carat using seaborn
sns.lmplot(x='carat', y='price', data=diamonds, line_kws={'color': 'red'}, height=7)
plt.title('Linear Regression: Price vs Carat')
plt.xlabel('Carat')
plt.ylabel('Price')

# Calculating linear regression line and associated statistics
slope, intercept, r_value, p_value, std_err = stats.linregress(diamonds['carat'], diamonds['price'])
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
print("P-value:", p_value)
print("Standard Error:", std_err)

plt.show()
```

**What do you think?**

```python
#there is a linear correlation tendency but a lot of deviation still
```

**Would you do any other checks on other features? Do you have any comments regarding `carat`?**

```python
# Plotting linear regression between price and x using seaborn
sns.lmplot(x='x', y='price', data=diamonds, line_kws={'color': 'red'}, height=7)
plt.title('Linear Regression: Price vs X')
plt.xlabel('X')
plt.ylabel('Price')

# Calculating linear regression line and associated statistics for x
slope_x, intercept_x, r_value_x, p_value_x, std_err_x = stats.linregress(diamonds['x'], diamonds['price'])
print("For X:")
print("Slope:", slope_x)
print("Intercept:", intercept_x)
print("R-squared:", r_value_x**2)
print("P-value:", p_value_x)
print("Standard Error:", std_err_x)

plt.show()

# Plotting linear regression between price and y using seaborn
sns.lmplot(x='y', y='price', data=diamonds, line_kws={'color': 'green'}, height=7)
plt.title('Linear Regression: Price vs Y')
plt.xlabel('Y')
plt.ylabel('Price')

# Calculating linear regression line and associated statistics for y
slope_y, intercept_y, r_value_y, p_value_y, std_err_y = stats.linregress(diamonds['y'], diamonds['price'])
print("For Y:")
print("Slope:", slope_y)
print("Intercept:", intercept_y)
print("R-squared:", r_value_y**2)
print("P-value:", p_value_y)
print("Standard Error:", std_err_y)

plt.show()
```

```python
#x,y (and z) is also impactful. Hence, i try to model price using these 3 factors:
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Selecting features (independent variables) and target (dependent variable)
X = diamonds[['carat', 'x', 'y']]  # Features: carat, x, y
y = diamonds['price']  # Target: price

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Creating and fitting the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Making predictions on the testing set
y_pred = model.predict(X_test)

# Calculating the Mean Squared Error (MSE) as a measure of model performance
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

# Printing the coefficients of the model
print("Coefficients:", model.coef_)

# Printing the intercept of the model
print("Intercept:", model.intercept_)

# Calculating Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python
#the model is not fit, suggesting a non-linear relationship
```
