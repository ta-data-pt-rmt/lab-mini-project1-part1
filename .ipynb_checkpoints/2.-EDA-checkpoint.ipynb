# 4. Taking a deeper look with plots and correlations

To take deeper look, we will use the `pairplot` method of `seaborn` library. This method plots a scatterplot for each pair of features and in the diagonal the distribution of the feature.

So if you have many features it will take a while, be careful!


```python
#Run this code
import seaborn as sns
sns.pairplot(diamonds, diag_kind = 'kde', plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
```

**What do you see here? What relationships between variables are the most interesting?**

Now we will see a correlation matrix with a plot. As you know a higher correlation means that the feature could be an effect (**but is not for sure**) for the changes on the price.

We will see this with a matrix with colors. A lighter color means greater correlation. 

This is done with the `seaborn` library as well.

```python
# your code here

sns.heatmap(diamonds.corr())
```

Finally, we will calculate the linear regression between the price and the weight. This will be done first by plotting it with the `seaborn` library and then calculating the error with the `scipy` library.

```python
# your code here

sns.set_style('whitegrid')
display(sns.lmplot(x ='price', y ='carat', data = diamonds, line_kws={"color": "yellow"}))

display(f"Carat std error: {stats.sem(diamonds['carat'])}")
display(f"Price std error: {stats.sem(diamonds['price'])}")
```

**What do you think?**

```python
#your thoughts here

'''Carat has a very low standard error, this means that the data of carat is concentrated around the mean. Price has a high standard error, so the data is much more spread out. We can decrease this standard error if we increase the size of the size of the sample '''
```

**Would you do any other checks on other features? Do you have any comments regarding `carat`?**

```python
#your thoughts here

display(f"color std error: {stats.sem(diamonds['color'])}")
display(f"clarity std error: {stats.sem(diamonds['clarity'])}")
display(f"depth std error: {stats.sem(diamonds['depth'])}")
display(f"table std error: {stats.sem(diamonds['table'])}")
display(f"x std error: {stats.sem(diamonds['x'])}")
display(f"y std error: {stats.sem(diamonds['y'])}")
display(f"z std error: {stats.sem(diamonds['z'])}")

'''Analyzing the graphs of the pairplot, it seems that no other features have very spread out the data, I can test them anyway. Carat (weight) is reasonable that be related with the dimensions, the heavier the larger '''
```
